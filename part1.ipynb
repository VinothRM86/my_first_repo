```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("medical_insurance.csv")
df.head()
# Display the number of rows and columns
print(f"Dataset Shape: {df.shape}")

# Display data types of all columns
print("Data Types:\n")
print(df.dtypes)

# Calculate null value count for each column
null_count = df.isnull().sum()

# Calculate null value percentage
null_percentage = (null_count / len(df)) * 100

# Create a summary
null_summary = pd.DataFrame({
    "Null Count": null_count,
    "Null Percentage (%)": null_percentage.round(2)
})

print("===== Null Value Summary =====")
display(null_summary)

# Identify columns with more than 20% null values
high_null_columns = null_summary[
    null_summary["Null Percentage (%)"] > 20
]

print("===== Columns with More Than 20% Null Values =====")
if high_null_columns.empty:
    print("No columns have more than 20% missing values.")
else:
    display(high_null_columns)

# Select numeric columns
numeric_columns = df.select_dtypes(include='number').columns

print("===== Median Computation =====")

for col in numeric_columns:

    # Calculate null percentage for the current column
    null_percentage = (df[col].isnull().sum() / df.shape[0]) * 100

    # Fill only if null values exist and percentage is less than 20%
    if 0 < null_percentage < 20:

        median_value = df[col].median()

        df[col] = df[col].fillna(median_value)

        print(f"{col:<12} --> Null values filled with median ({median_value:.2f})")
    
print("===== Null Values After computation =====")
remaining_null = df.isnull().sum()
display(remaining_null)

# Count duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# df shape before removing duplicates
rows_before = df.shape[0]
print(f"Rows before removing duplicates: {rows_before}")

# Remove duplicate rows
df = df.drop_duplicates()
rows_after = df.shape[0]
print(f"Rows after removing duplicates: {rows_after}")

#number of removed rows
rows_removed = rows_before - rows_after
print(f"Number of duplicate rows removed: {rows_removed}")

# Null value percentage after duplicate removal
null_percentage_after = (
    df.isnull().sum() / df.shape[0]
) * 100
null_percentage_after = null_percentage_after.round(2)

#Comarison of null value percentage before and after duplicate removal
print("Null value percentage comparison")
comparison = pd.DataFrame({
    "Before Duplicate Removal (%)": null_percentage.round(2),
    "After Duplicate Removal (%)": null_percentage_after
})
display(comparison)

## Task 4: Data Type Correction
print("Current Data Types:\n")
print(df.dtypes)

#memory usage before conversion
memory_before = df.memory_usage(deep=True).sum()
print(f"Memory Usage Before Conversion: {memory_before:,} bytes")

# Simulate an incorrect data type
df["children"] = df["children"].astype("object")
print(df.dtypes)
#covert an incorrect data type to numeric
df["children"] = pd.to_numeric(df["children"], errors='coerce')
print(df.dtypes)
#Convert object columns to category type
categorical_columns = ["sex", "smoker", "region"]
for col in categorical_columns:
    df[col] = df[col].astype("category")
print(df.dtypes)

#memory usage after conversion
memory_after = df.memory_usage(deep=True).sum()
print(f"Memory Usage After Conversion: {memory_after:,} bytes")

#Task 5: Descriptive statistics and skewness
# Generate descriptive statistics for all numeric columns
numeric_df = df.select_dtypes(include='number')
print("===== Descriptive Statistics =====")
display(numeric_df.describe())

#Calculate skewness
print("===== Skewness of Numeric Columns =====")
skewness = numeric_df.skew()
display(skewness)

# Find the column with the highest absolute skewness
highest_skew_column = skewness.abs().idxmax()
highest_skew_value = skewness[highest_skew_column]

print(f"Column with highest absolute skewness: {highest_skew_column}")
print(f"Skewness Value: {highest_skew_value:.2f}")

#Task 6: Outlier Detection with IQR method
def detect_outliers_iqr(df, column):

    # Calculate Q1 and Q3
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)

    # Calculate IQR
    IQR = Q3 - Q1

    # Calculate lower and upper bounds
    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    # Identify outliers
    outliers = df[
        (df[column] < lower_bound) |
        (df[column] > upper_bound)
    ]

    # Print summary
    print(f"\n===== {column.upper()} =====")
    print(f"Q1           : {Q1:.2f}")
    print(f"Q3           : {Q3:.2f}")
    print(f"IQR          : {IQR:.2f}")
    print(f"Lower Bound  : {lower_bound:.2f}")
    print(f"Upper Bound  : {upper_bound:.2f}")
    print(f"Outliers     : {len(outliers)}")

    return outliers

#BMI outliers detection
bmi_outliers = detect_outliers_iqr(df, "bmi")
#Charges outliers detection
charges_outliers = detect_outliers_iqr(df, "charges")
#Outliers count
outlier_count = pd.DataFrame({
    "Column": ["bmi", "charges"],
    "Outlier Count": [
        len(bmi_outliers),
        len(charges_outliers)
    ]
})
display(outlier_count)

#Task 7: Data Visualization
# Line plot for charges
plt.figure(figsize=(8,5))
plt.plot(df.index, df["charges"])
plt.title("Medical Insurance Charges by Row Index")
plt.xlabel("Row Index")
plt.ylabel("Insurance Charges")

plt.show()
#Bar chart for smoker status
mean_charges = df.groupby("smoker")["charges"].mean()

plt.figure(figsize=(8,5))

plt.bar(mean_charges.index.astype(str), mean_charges.values)

plt.title("Average Insurance Charges by Smoker Status")
plt.xlabel("Smoker")
plt.ylabel("Average Charges")

plt.show()

#Histogram for the most skewed column (charges)
plt.figure(figsize=(8,5))
sns.histplot(df[highest_skew_column], bins=20)
plt.title(f"Distribution of {highest_skew_column}")
plt.xlabel(highest_skew_column)
plt.ylabel("Frequency")

plt.show()

#Scatter plot for bmi vs charges
plt.figure(figsize=(8,5))

sns.scatterplot(
    x="bmi",
    y="charges",
    data=df
)

plt.title("BMI vs Insurance Charges")
plt.xlabel("BMI")
plt.ylabel("Charges")

plt.show()

#Box plot for charges by smoker status
plt.figure(figsize=(8,5))

sns.boxplot(
    x="smoker",
    y="charges",
    data=df
)
plt.title("Insurance Charges by Smoker Status")
plt.xlabel("Smoker")
plt.ylabel("Charges")

plt.show()

#Task 8: Correlation Analysis
# Compute correlation matrix
correlation_matrix = numeric_df.corr()
print("===== Correlation Matrix =====")
display(correlation_matrix)
#Create heatmap for correlation matrix
plt.figure(figsize=(8,5))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)
plt.title("Pearson Correlation Heat Map")

plt.show()

#Identify highest absolute correlation pair
# Remove self-correlations
corr_without_diagonal = correlation_matrix.copy()

np.fill_diagonal(corr_without_diagonal.values, 0)

# Find the highest absolute correlation
highest_pair = corr_without_diagonal.abs().stack().idxmax()
highest_value = corr_without_diagonal.loc[highest_pair]

print(f"Highest Absolute Correlation Pair : {highest_pair}")
print(f"Correlation Value : {highest_value:.2f}")

#Task 9: Imputation strategy comaparison
original_df = pd.read_csv("medical_insurance.csv")
# Identify most skewed columns
top_skewed_columns = skewness.abs().sort_values(ascending=False).head(2).index

print("Top 2 Most Skewed Columns:")
print(top_skewed_columns.tolist())

#Compare mean and median
comparison_list = []

for col in top_skewed_columns:

    comparison_list.append({
        "Column": col,
        "Mean": original_df[col].mean(),
        "Median": original_df[col].median(),
        "Skewness": skewness[col]
    })

comparison_df = pd.DataFrame(comparison_list)

display(comparison_df)

#Median imputation for skewed columns
for col in top_skewed_columns:

    original_df[col] = original_df[col].fillna(
        original_df[col].median()
    )
#Confirmation of no null values
print(original_df[top_skewed_columns].isnull().sum())

#Spearman rank correlation for skewed columns
display(correlation_matrix)
spearman_matrix = numeric_df.corr(method="spearman")
print("Spearman Correlation Matrix")
display(spearman_matrix)

#Difference between Pearson and Spearman correlation
print("Difference between Pearson and Spearman Correlation")
difference_table = (
    (spearman_matrix - correlation_matrix)
    .abs()
)

display(difference_table)

#Grouped aggregation:
group_summary = (
    df.groupby("smoker")["charges"]
      .agg(["mean", "std", "count"])
)

display(group_summary)
#Highest Mean:
highest_mean_group = group_summary["mean"].idxmax()
print("Highest Mean Group:", highest_mean_group)
#Highest Standard Deviation:
highest_std_group = group_summary["std"].idxmax()
print("Highest Standard Deviation Group:", highest_std_group)
#Mean Ratio:
mean_ratio = (
    group_summary["mean"].max()
    /
    group_summary["mean"].min()
)

print(f"Mean Ratio: {mean_ratio:.2f}")

#Task 10: Save the cleaned dataset
df.to_csv("cleaned_data.csv", index=False)
print("✅ cleaned_data.csv has been saved successfully.")

```
